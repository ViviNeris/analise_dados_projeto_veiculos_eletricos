```python
# Importando a biblioteca Pandas
import pandas as pd
```

```python
# Criar uma variável "tabela" que irá armazenar a base, codificando para América latina
tabela = pd.read_csv("/content/drive/MyDrive/Data Science/IEA Global EV Data 2024.csv", encoding="latin1")
```

```python
# Exibir tabela
display (tabela)
```

```python
# Atualizar a tabela eliminando a linha sem dados
tabela = tabela.dropna()

# Exibindo as informações da minha tabela
display(tabela.info())
```

```python
# Exibindo a descrição dos dados da Tabela
display(tabela.describe().round(1))
```

```python
# Importando Biblioteca Plotly Express
import plotly.express as px

# Selecionando apenas as linhas que tem o parâmetro 'EV Sales'
vendas_ano = tabela[tabela['parameter'] == 'EV sales']

# Agrupar por ano e somar os valores de vendas. Usando 'groupby'para agrupar os dados da coluna 'year'. E usando o .sum para somar os valores da coluna'value'
# Nesse caso analisamos Quantidades Vendidas x Anos
vendas_ano_agrupadas = vendas_ano.groupby('year')['value'].sum().reset_index() # Reset.index para agrupar os dados de forma organizada

# Criar o gráfico interativo com Plotly Express
fig = px.line(vendas_ano_agrupadas, x='year', y='value',
              title="Evolução das Vendas de Veículos Elétricos ao Longo dos Anos",
              labels={'value': 'Quantidade de Vendas', 'year': 'Ano'})

# Mostrar o gráfico
fig.show()
```

O gráfico mostra uma estabilidade nas vendas de veículos elétricos até mais ou menos o ano de 2018, e após isso há um crescimento exponencial. Esse aumento pode ser atribuído a fatores como avnanço tecnológicos e maior conscientização quanto ao meio ambiente. As projeções para 2030 a 2025 indicam a continuação desse crescimento.
